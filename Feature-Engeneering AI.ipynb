```python
'''
Q-3> Define Machine Learning. What are the main components in Machine Learning ?
Sol:
Machine Learning is a branch of artificial intelligence (AI) that allows computers to learn patterns from data and make decisions or predictions 
without being explicitly programmed for every task.
Instead of giving the computer rules to follow, we give it data, and it figures out the rules on its own.
Main Components of Machine Learning:
Data:
1.The foundation of ML.
2.Can be images, text, numbers, audio, etc.
Split into:
1.Training data (used to learn)
2.Testing data (used to evaluate)
Features:
1.The input variables used to make predictions.
ex: In house pricing, features might be size, location, number of rooms.
Model:
1.The mathematical representation of the relationship between features and output.
Ex: Linear regression, decision trees, neural networks.
Algorithm:
1.The method used to train the model.
2.It tells the model how to learn from the data.
Ex: Gradient descent, backpropagation, k-nearest neighbors
Training:
1.The process where the model learns patterns from the data.
2.Adjusts internal parameters (like weights in a neural network) to minimize error.
Prediction / Inference:
1.After training, the model makes predictions on new (unseen) data.
Evaluation:
1.Measures how well the model performs.
2.Common metrics: Accuracy, precision, recall, F1 score, mean squared error
Hyperparameters:
1.Settings that control the learning process (not learned from data).
Ex: Learning rate, number of trees, batch size.
'''
```

```python
'''
Q-6> How do we handle categorical variables in Machine Learning? What are the common techniques?
Sol:
Machine learning models need numbers to process input. So we need to convert categories into numerical format while still preserving the
meaning or structure.
Label Encoding:
Color	Encoded
Red	       0
Blue	   1
Green	   2
One-Hot Encoding:
Color	Red	      Blue	Green
Red	      1	       0	 0
Blue	  0	       1	 0
Frequency Encoding:
Replace categories with their frequency (or count) in the dataset.
Target Encoding (Mean Encoding):
Replaces each category with the mean value of the target for that category.
Category	Average Target
A	             0.75
B	             0.52
Binary Encoding / Hashing Encoding:
1.Used for high-cardinality features (lots of categories).
2.Reduces dimensionality compared to one-hot.
Situation	                               Best Choice
Few categories                       	One-hot encoding
Many categories	                        Frequency or target encoding
Tree-based models	                    Label or target encoding
Linear/NN models	                    One-hot or binary encoding
'''
```

```python
'''
Q-7> What do you mean by training and testing a dataset ?
Sol:
Training Dataset:
1.This is the data the model learns from.
2.It contains input features (like height, age) and the correct output (label).
3.The model looks at this data, tries to understand patterns, and adjusts its internal parameters (like weights in a neural network).
Testing Dataset:
1.This is new, unseen data used to evaluate how well the model performs.
2.You never show this data during training.
3.It helps you see whether the model generalizes well to data it hasn’t seen before.
'''
```

```python
'''
Q-11> Why do we have to perform EDA before fitting a model to the data ?
Sol:
Understand the Data:
1.EDA helps you gain insights into your data, which is crucial for:
2.Understanding which features are relevant to your model.
3.Knowing the distribution of variables (e.g., continuous vs. categorical).
4.Understanding the relationships between features (e.g., correlations, patterns).
5.Without EDA, you might miss critical insights and end up with an ineffective model.
Handle Missing or Incorrect Data :
1.Many datasets contain missing values, outliers, or incorrect entries that can skew model performance.
2.Missing Values: You might decide to impute them (fill with the mean/median) or drop rows/columns.
3.Outliers: EDA lets you spot outliers that could affect the model.
Identify Data Types & Feature Engineering :
EDA shows:
1.Which features are numerical vs. categorical.
2.How to transform categorical variables (e.g., using one-hot encoding or label encoding).
3.Whether scaling is necessary (e.g., for models that are sensitive to feature scaling like SVM or logistic regression).
Detect Class Imbalances :
1.For classification problems, EDA helps you detect if your target variable is imbalanced (e.g., if 90% of your labels are of one class and 10% are another).
2.Why does this matter? If the model sees many more examples of one class, it may become biased toward predicting that class, affecting performance on the minority class.
3. After that it decide whether to use techniques like resampling or class weights.
Understand Feature Relationships :
1.EDA allows you to explore the relationships between features and the target variable:
2.Correlation analysis helps you identify highly correlated features, which may cause multicollinearity.
3.Visualization (scatter plots, pair plots, etc.) shows how features interact with each other.
Choose the Right Model :
Not all models work the same way for every type of data. For instance:
1.If you have a linear relationship between features, a linear model (like Linear Regression) might be a good fit.
2.If the relationship is non-linear, you might opt for tree-based models like Decision Trees or Random Forests.
Determine the Need for Data Transformation :
Certain ML models require that the data be transformed:
1.Normalization/Standardization might be needed for models like SVM, kNN, or logistic regression.
2.Log transformation can help with highly skewed data (for example, income data).
Detect Data Leakage Risks :
1.Data leakage happens when information from outside the training dataset is used to create the model. It artificially inflates the performance of your model and can cause it to fail when deployed in real-world situations.
2.EDA helps to spot potential leakage sources (like including a feature that’s too closely related to the target) and correct them early.
'''
```

```python
'''
Q-15> What is causation? Explain difference between correlation and causation with an example ?
Sol:
Causation:
Causation means that one variable directly affects or causes a change in another.
1.If X causes Y, then changing X will lead to a change in Y.This is a cause-and-effect relationship.
Correlation:
Correlation means that two variables are related or move together, but it doesn’t necessarily mean one causes the other.
If X and Y are correlated, they change together—but X might not cause Y.
Difference Between Correlation and Causation:
Aspect	                                              Correlation	                                                Causation
Relationship	                                Two variables move together	                        One variable causes the other to change
Direction	                                    Can be positive or negative	                        Typically has a clear direction of effect
Guarantees Cause	                                No guarantee	                                          Yes, by definition
Proved by	                                    Statistical analysis (e.g., Pearson's r)	         Controlled experiments, domain knowledge
'''
```

```python
'''
Q-16> What is an Optimizer? What are different types of optimizers? Explain each with an example ?
Sol:
An optimizer is an algorithm that adjusts the model’s weights during training to minimize the loss function 
(i.e., reduce the error between predictions and actual values).
Types:
1.Gradient Descent (GD)
Basic idea: Move in the direction that reduces the loss the most (i.e., the negative gradient)
Formula:
θ = θ - α * ∇L
2.Stochastic Gradient Descent (SGD)
3.Mini-Batch Gradient Descent
4.Momentum
5.RMSprop (Root Mean Square Propagation)
6.Adam (Adaptive Moment Estimation) : Most popular optimizer in deep learning.
It combines:
1.Momentum (for smoothing)
2.RMSprop (for adaptive learning rates)
Ex:
'''
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.optimizers import Adam
X_train = np.random.rand(100, 10)            
y_train = np.random.randint(0, 2, size=(100, 1))  
model = Sequential([Input(shape=(10,)),Dense(32, activation='relu'),Dense(1, activation='sigmoid')])
model.compile(optimizer=Adam(learning_rate=0.001),loss='binary_crossentropy',metrics=['accuracy'])
history = model.fit(X_train, y_train, epochs=5, batch_size=16, verbose=1)
X_test = np.random.rand(5, 10)               
predictions = model.predict(X_test)
print("\nPredictions on new data:")
print(predictions)
```
